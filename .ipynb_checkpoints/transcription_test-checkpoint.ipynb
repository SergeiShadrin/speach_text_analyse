```python
# to transform from video to audio


def video_to_audio(video, audio):
    ffmpeg_cmd = [
        "ffmpeg", "-y",
        "-hide_banner",
        "-loglevel", "error",
        "-i", video,
        "-vn",
        "-ac", "1",
        "-ar", "16000",
        "-acodec", "pcm_s16le",
        audio
    ]
    try:
        subprocess.run(ffmpeg_cmd, check=True, text=True, capture_output=True)
    except subprocess.CalledProcessError as e:
        print("FFmpeg command:", " ".join(ffmpeg_cmd))
        print("STDERR:\n", e.stderr)
        raise
video_to_audio("data/input/test_01.mp4", "data/input/audio/test_01.wav")
```

```python
# chunking logic implementation
from pydub import AudioSegment

song = AudioSegment.from_mp3("data/input/audio/test_01.wav")

# PyDub handles time in milliseconds
ten_minutes = 10 * 60 * 100

first_10_minutes = song[:ten_minutes]

first_10_minutes.export("data/input/audio/test_01_10.wav", format="wav")
```
