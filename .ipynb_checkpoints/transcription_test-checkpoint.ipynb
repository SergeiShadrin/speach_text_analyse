```python
import os
import subprocess
from app.models.media_converter import FFmpegMediaConverter
from app.models.media_chunker import AudioChunker
from app.models.opanai_transcriber import OpanAITranscriber

from dotenv import load_dotenv
import replicate
import pprint

load_dotenv() 
```

```python
converter = FFmpegMediaConverter()
audio_path = converter.extract_audio("data/input/test_01.mp4")
print(audio_path)
```

```python
chunker = AudioChunker()
chunked_audios = chunker.split(audio_path)
print(chunked_audios)
```

```python
transcriber = OpanAITranscriber()


transcription = ""
for item in chunked_audios:
    text_item = transcriber.transcribe(
        input_path = item,
        model="gpt-4o-transcribe-diarize", 
        response_format="diarized_json", 
        language="fr", 
        prompt="les Assises nationales de la recherche partenariale organisées par ANRT - Association Nationale de la Recherche et de la Technologie en français."
        )
    print(text_item)
    
    transcription += text_item

filename = os.path.basename(input_path)
_name = os.path.splitext(filename)[0]
output_path = f"data/output/transcription/{_name}.wav"
with open(output_path, "w") as f:
  f.write(f"{transcription}")
```

```python

transcription = ""
for item in chunked_audios:
    text_item = transcriber.transcribe(
        input_path = item,
        model="whisper-1", 
        language="fr", 
        prompt="les Assises nationales de la recherche partenariale organisées par ANRT - Association Nationale de la Recherche et de la Technologie en français."
        )
    print(text_item)
    
    transcription += text_item

```

```python
filename = os.path.basename(audio_path)
_name = os.path.splitext(filename)[0]
output_path = f"data/output/transcription/{_name}.txt"
with open(output_path, "w") as f:
  f.write(f"{transcription}")
```
